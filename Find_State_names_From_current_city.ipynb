```python
test_location_df = pd.read_csv('dataset/Permnent_address_test_location.csv')
print(test_location_df.shape)
test_location_df.columns
```

```python
test_location_df['Location'] = test_location_df['Location'].str.lower()

test_location_df['Location'] = test_location_df['Location'].str.replace('[#,@,&,(,),.,/,;]', ' ').dropna()

test_location = test_location_df['Location'].str.replace(',', ' ')
```

```python
key = ['MAHARASTRA', 'KARNATAKA', 'ODISHA', 'ANDHRA PRADESH', 'WEST BENGAL', 'MADHYA PRADESH', 'TAMIL NADU', 'UTTAR PRADESH', 'RAJASTHAN', 'PUNJAB', 'KERALA', 'BIHAR', 'HARYANA', 'HIMACHAL PRADESH', 'ARUNACHAL PRADESH', 'ASSAM', 'GUJARAT', 'DELHI', 'CHHATTISGARH', 'GOA', 'TRIPURA', 'JHARKHAND']
key_lower = [e.lower() for e in key]
```

```python
def Test_Location(review):
    state_loc = []
    review = str(review).strip()
    if review in key_lower:
        state = get_key(review)
        return state
    for token in review.split():
        token = ''.join(e.lower() for e in token if e.isalnum())
        token = token.strip()
        state = get_key(token)
        state_loc.append(state)
    return ' '.join(state_loc)
```

```python
count = 0
state_name_list = []
file_path = "State_output.csv"
if not os.path.exists(file_path):
    df = pd.DataFrame(columns=['current_Location', 'State_location'])
    df.to_csv('State_output.csv')
    print("File Created Successfully...!")
    
test_df = pd.read_csv('State_output.csv', index_col=[0])   

for _, item in test_location_df.iterrows():
    state_names = Test_Location(item['Location'])
    if "key doesn't exist" in state_names:
        state_names = state_names.replace("key doesn't exist", '').strip()
        for state_name in state_names.split():
            if state_name not in state_name_list:
                state_name_list.append(state_name)
        if len(state_name_list) > 0:
            state_list_value = state_name_list[-1]
            print(item['Location'],'*********',state_list_value)
            test_df = test_df.append(pd.Series([item['Location'],state_list_value], index=df.columns), ignore_index=True)
        else:
            print("no state found...",item['Location'])
            test_df = test_df.append(pd.Series([item['Location'],'state_not_found'], index=df.columns), ignore_index=True)
            count = count +1
        state_name_list = []
    else:
        print(item['Location'],'*********',state_names)
        state_names = state_names.split()[-1]
        test_df = test_df.append(pd.Series([item['Location'],state_names], index=df.columns), ignore_index=True)
print("---------Count of State Not found Records...",count)
test_df.to_csv('State_output.csv')
```
